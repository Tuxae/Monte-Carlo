### Some useful theorems :

* Law of Large Numbers : Let $(X_n)_{n \geq 1}$ a sequence of i.i.d. Lebesgue integrable random variables. Then as $n$ approaches infinity, $$\frac{1}{n}\sum_{k=1}^n X_k \overset{\mathcal{a.s.}}{\underset{n \to +\infty}{\longrightarrow}} E[X_1]$$.

* Central Limit Theorem : Let $(X_n)_{n \geq 1}$ a sequence of i.i.d. random variables with $E[X_i] = \mu$ and $Var[X_i] = \sigma^2 <\infty$. Then as $n$ approaches infinity, the random variables $$\sqrt{n}(\bar{X_n} âˆ’ \mu) \overset{\mathcal{L}}{\underset{n \to +\infty}{\longrightarrow}} \mathcal{N}(0,\sigma^2)$$.

## Illustrations :

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Law of Large Numbers :

```python
n, p  = 1, 1/3
sizes = [10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 5*10**7]

for i, size in enumerate(sizes) :
    X = np.random.binomial(n, p, size=size)
    print("#" + str(i) + " : " + str(X.mean()))
```

#### Central Limit Theorem :

```python
# beta = 1 / lambda 
# => beta =  expected value
beta = 0.1 
sizes = [10, 10**3, 10**5, 10**6]

for i, size in enumerate(sizes):
    X = [np.random.exponential(beta, size=size).mean() for i in range(1000)]
    plt.figure(figsize=(6,6))
    plt.ylabel("Frequency")
    plt.xlabel("Distribution of the expected value")
    plt.hist(X, bins=30)
```

## Pseudo Random Number Generators

1948 : linear congruential generators (Derrick Lehmer).

$$ X_{n+1} = (a X_n + c) \,\, (mod \, m)$$

and $X_0$ is the seed.

Critical choices of $a, c$ and $m$ :

```python
a = 25
c = 16
m = 256
seed = 50
n = 20

X = [seed]
for i in range(1,n):
    X.append((a*X[i-1]+c)%m)

print(X)
```

1982 : Wichmann-Hill by Brian Wichmann and David Hill

```python
# Need to set nb1, nb2, nb3 before
# between 1 and 30000

nb1, nb2, nb3 = 5, 2344, 9645
```

```python
def Wichmann_Hill(nb1, nb2, nb3):
    n1 = (171 * nb1) % 30269
    n2 = (172 * nb2) % 30307
    n3 = (170 * nb3) % 30323
    
    return (n1/30269 + n2/30307 + n3/30323) % 1
```

```python
Wichmann_Hill(nb1, nb2, nb3)
```
